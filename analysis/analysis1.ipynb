The box plot will provide information between income category and credit limit. Some bckground information relevant to understand credit limits better:
Credit limits are indicators that allow banks and financial institutions to depict the amount of money or credit they are willing to offer a borrower as loan. When consumers have higher credit scores, they are more favourable for loans. 
One assumption taken in this boxplot is that there is a connection between the 2 variables. Consumers with higher incomes will have better credit limits as they are more likely to pay off the debt sooner. 
As shown in the graph below, we can see that predicition we had for higher incomes having credit limits holds true. The median in the boxplot( the line cutting the box) is the highest for incomes greater than 120k. The outlers in the box plots shows credit limit for some consumers that is not appropriate for their income category. For example the category of 60k-80k has high credit limit outliers which can be due to consumers having shown bank loyalty and the bank knows and approves of the consumer's banking practices even though they have lower incomes.  

```python
ax1=sns.boxplot(x='Income_Category',y='Credit_Limit',data=df)
ax1.set(title='Relationship between income and credit',
        xlabel='Income category',
        ylabel='Credit limit')
ax1.set_ylim(bottom=0)
```

As I mentioned before in my research question, I will be adding data from Pew Research Center. This dataset talks about the percentage of mobile banking and internet usage for banking activities. This is important as through this we will understand the relationship between online banking and credit card churning. 
One assumption I have used is that more advancements in technology means that more people are convertnig to online banking. 

```python
data = pd.DataFrame({'% of cell phone users who do mobile banking': [27,31,34,35]},
                  index=['Less than 30k', '30k-49k','50k-74k','Greater than 75k'])
data
```

```python
plot = data.plot.pie(y='% of cell phone users who do mobile banking', figsize=(11, 6))
plot.legend(loc='upper right')
```

As the pie chart shows that there is not a significant difference in the % of people doing online banking. However, there is still an increasing trend of people using mobile banking as income category increases. Red shaded area is a little larger than the rest. 
Since we have found that more consumers with a higher income category perform more online banking, we can draw a correlation between consumers and credit card churning based on their income category. 
The following graph shows the relationship between income category and consumer retention.Consumer retention is shown by the months on book column which shows the period of relationship with the bank.

```python
plot3=sns.violinplot(x="Income_Category",y="Months_on_book", data=df2)
plot3.set(xlabel='Income',
        ylabel='Period of Relationship with bank', 
        title= "Relationship between length of stay of consumers and income")
```
