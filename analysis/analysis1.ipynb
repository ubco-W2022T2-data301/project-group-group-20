Now this looks like a readable dataset with columns easy to understand! Another column that seems important to the research question is a column that categorize months on book into low, medium and high categories. This will make readers easily understand if consumers are more likely to participate in credit card churning. Lower the months on book, the consumers are more likely to paricipate in credit card churning. The fourth step to process data will be to replace "Unknown" values to become "Not Specified". Last step to process data is to deal with outliers. I will be focusing on outliers in months on book column as well as the credit limit column. One way to deal with them is finding out the median of the values.

```python
def categorize_period_of_relationship(i):
    if i <= 12:
        return 'low'
    elif i <= 24:
        return 'medium'
    else:
        return 'high'
df_cleaned['MonthsOnBookCategory'] = df_cleaned['Months_on_book'].apply(categorize_period_of_relationship)
df_cleaned['Education_Level'] = df_cleaned['Education_Level'].replace('Unknown', 'Not specified')
df_cleaned['Marital_Status'] = df_cleaned['Marital_Status'].replace('Unknown', 'Not specified')
df_cleaned
```

<b> Analysis </b>

Credit card churning is a practice done by consumers where they sign up for rewards and bonuses from multiple credit cards and cancel them as soon as the annual fees begin. Technology has been an important factor in making credit card churning easier as with the help of mobile banking, one doesnt have to go to banks to sign up for rewards only to cancel the card a month later. Not only that, but through online banking, consumers can also easily compare different card offers, track their rewards and set reminders to cancel them. They can all do these various activities on their mobiles, making this way faster and easier. 
However, it is only the advancements in technology have the banks started using this to their advantage.Banks themeselves are using advanced alogrithms and machines to identify potential churners and flag them before they continue this practice.For example, if a consumer has signed up for mulitple credit cards within few months into being associated with the bank, it shows a potential churner. 
Banks have also used technological advancements in their advantage by making it harder for consumers to chrun their cards. One way this is done is by making consumers wait a certain time before they are issued another card and also limit the number of cards issued by a consumer at a time. Banks have also added filters and categories that have to be met in order to issue credit cards. For example, a premium credit card issuer must be educated, and have an income category of over 120k. These thresholds discourage churning.
